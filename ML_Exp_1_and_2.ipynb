```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/drive/MyDrive/ML Exp 1/Consumer_Price_Index_in_Denver.csv'
df = pd.read_csv(file_path)

# Filter for Annual CPI-U values
filtered_df = df[(df['periodTypeDescription'] == 'Annual') & (df['title'].str.contains('CPI-U'))]

# Sort by year
filtered_df = filtered_df.sort_values('periodYear')

# Plot
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['periodYear'], filtered_df['cpi'], marker='o', linestyle='-', color='teal')
plt.title('Annual CPI-U in the US Over Years')
plt.xlabel('Year')
plt.ylabel('Consumer Price Index')
plt.grid(True)
plt.tight_layout()
plt.show()
```
